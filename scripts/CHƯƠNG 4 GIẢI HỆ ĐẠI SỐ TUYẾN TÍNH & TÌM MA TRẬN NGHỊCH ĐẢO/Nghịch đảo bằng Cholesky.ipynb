### Nghịch đảo qua Cholesky


```python
import numpy as np
from IPython.display import display, Math, Markdown

def _mat(M, p=4):
    """Chuyển đổi ma trận hoặc mảng NumPy thành chuỗi LaTeX bmatrix để hiển thị."""
    if hasattr(M[0], "__len__"):
        rows = " \\\\ ".join([" & ".join([f"{x:.{p}f}" for x in row]) for row in M])
        return f"\\begin{{bmatrix}} {rows} \\end{{bmatrix}}"
    inner = " \\\\ ".join([f"{x:.{p}f}" for x in M])
    return f"\\begin{{bmatrix}} {inner} \\end{{bmatrix}}^T"

def nghich_dao_Cholesky(A_input):
    """
    Tìm ma trận nghịch đảo A^-1 thông qua phân tách Cholesky (A = L * L^T).
    Giải hệ phương trình L * L^T * X = I theo từng cột.
    """
    A = np.array(A_input, dtype=float)
    n = A.shape[0]
    
    # -----------------------------------------------------------------
    # GIAI ĐOẠN 1: KHỞI TẠO VÀ HIỂN THỊ MA TRẬN ĐẦU VÀO
    # -----------------------------------------------------------------
    display(Markdown("## ❖ TÌM MA TRẬN NGHỊCH ĐẢO $A^{-1}$ QUA PHÂN TÁCH CHOLESKY"))
    display(Markdown("---"))
    display(Markdown("### **1. Ma trận hệ số đầu vào**"))
    display(Math(f"A = {_mat(A)}"))

    # -----------------------------------------------------------------
    # GIAI ĐOẠN 2: PHÂN TÁCH CHOLESKY (A = L * L^T)
    # -----------------------------------------------------------------
    L = np.zeros((n, n))
    for i in range(n):
        s = sum(L[i, k]**2 for k in range(i))
        val = A[i, i] - s
        if val <= 0:
            display(Markdown("> ⚠️ **Cảnh báo:** Ma trận không xác định dương. Không thể áp dụng thuật toán Cholesky!"))
            return None
        L[i, i] = np.sqrt(val)
        for j in range(i + 1, n):
            L[j, i] = (A[j, i] - sum(L[j, k] * L[i, k] for k in range(i))) / L[i, i]

    display(Markdown("### **2. Kết quả phân tách Cholesky ($A = L L^T$)**"))
    display(Math(f"L = {_mat(L)} \\quad \\Rightarrow \\quad L^T = {_mat(L.T)}"))

    # -----------------------------------------------------------------
    # GIAI ĐOẠN 3: GIẢI HỆ PHƯƠNG TRÌNH TÌM MA TRẬN NGHỊCH ĐẢO
    # -----------------------------------------------------------------
    # Giải L * L^T * x_j = e_j với e_j là cột thứ j của ma trận đơn vị I
    X = np.zeros((n, n))
    for j in range(n):
        e = np.zeros(n)
        e[j] = 1.0
        
        # Bước 3.1: Giải hệ tiến L * y = e
        y = np.zeros(n)
        for i in range(n):
            y[i] = (e[i] - sum(L[i, k] * y[k] for k in range(i))) / L[i, i]
            
        # Bước 3.2: Giải hệ lùi L^T * x = y
        x = np.zeros(n)
        for i in range(n - 1, -1, -1):
            x[i] = (y[i] - sum(L[k, i] * x[k] for k in range(i + 1, n))) / L[i, i]
            
        X[:, j] = x

    # -----------------------------------------------------------------
    # GIAI ĐOẠN 4: HIỂN THỊ KẾT QUẢ VÀ KIỂM TRA ĐỘ CHÍNH XÁC
    # -----------------------------------------------------------------
    display(Markdown("---"))
    display(Markdown("### **3. Ma trận nghịch đảo kết quả**"))
    display(Math(f"A^{{-1}} = {_mat(X)}"))
    
    display(Markdown("### **4. Kiểm tra điều kiện nghiệm ($A \\cdot A^{-1} = I$)**"))
    display(Math(f"A \\cdot A^{{-1}} = {_mat(A @ X)}"))
    
    return X

# ═══════════════════════════════════════════════════════════════════
# KHU VỰC NHẬP DỮ LIỆU ĐẦU VÀO
# ═══════════════════════════════════════════════════════════════════
A = np.array([
    [4.0,  2.0,  2.0],
    [2.0,  5.0,  3.0],
    [2.0,  3.0, 10.0]
])
# ═══════════════════════════════════════════════════════════════════

# Thực thi chương trình
A_inverse = nghich_dao_Cholesky(A)
```
